```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
model = ResNet50(weights='imagenet')
```

```python
img_path = 'zebra.jpg'
#img_path = 'linjal.jpg'
#img_path = 'tennisboll.jpg'
img = image.load_img(img_path, target_size=(224, 224))
img
```

```python
x = image.img_to_array(img)
x.shape
```

```python
x = np.expand_dims(x, axis=0)
x.shape
```

```python
# https://www.tensorflow.org/api_docs/python/tf/keras/applications/resnet50/preprocess_input
x = preprocess_input(x)
```

```python
preds = model.predict(x)
# decode the results into a list of tuples (class, description, probability)
# (one such list for each sample in the batch)
print('Predicted:', decode_predictions(preds, top=3)[0])
# Predicted: [(u'n02504013', u'Indian_elephant', 0.82658225), (u'n01871265', u'tusker', 0.1122357), (u'n02504458', u'African_elephant', 0.061040461)]
```

# 2. Answer the question: Why is "Pretrained models" a very powerful concept?

Pretrained models are ready to use tools which includes knowleges.
They save time and resources by letting you build on what's already been
learned, making it easier to solve new problems or tasks without starting from scratch. You can save time and instead for building and training a new model can focus on data preprocessing and model adjusting, fine tuning instead.
Also pretrained models make less posibillity for errors of mistakes in code or formulas, because everythting is already created and checked.
It is like a sketch which every artist can make to the picture in their own way.
